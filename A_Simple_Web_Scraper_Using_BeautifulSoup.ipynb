```python
import requests
from bs4 import BeautifulSoup

page = 1
next_page = True

while next_page:
  
  r = requests.get('http://quotes.toscrape.com/page/'+ str(page))

  soup = BeautifulSoup(r.text, 'html5lib') 

  quote = soup.select('.quote')

  links = soup.find_all('a')

  next_page = soup.select_one('.next > a')

  print('The quotes mentioned in the website are the following:')
  print('--- ------ --------- -- --- ------- --- --- ---------')
  print(' ')
  for q in quote:
      txt = q.select_one('.text')
      author = q.select_one('.author')
      tag = q.select('.tag')
      print(txt.text)
      print(' ')
      print('--', author.text)
      print(' ') 
      print('Tags:')
      print('----')
      for n, j in enumerate(tag):
          print(' ')
          print('{})'.format(n), j.text) 
          print(' ')
      print('----------------------------------------------')
  print('Information retrived from page' + str(page))
  page += 1
  print('----------------------------------------------')
  print(' ')
```
